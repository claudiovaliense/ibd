### Algumas Visualizações

Podemos usar diversas bibliotecas gráficas para plotar gráficos a partir de dados reais ou inputs fixos.
Nestes exemplos, faremos algumas visualizações utilizando a biblioteca `matplotlib.pyplot` (https://matplotlib.org/api/_as_gen/matplotlib.pyplot.html)

```python
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
import math
```

```python
y_list = [1, 2, 3, 4]

plt.plot(y_list)
plt.ylabel('y')
plt.xlabel('x')
plt.show()
```

```python
x_list = [1, 4, 9, 16]
plt.plot(y_list, x_list)
plt.ylabel('y')
plt.xlabel('x')
plt.show()
```

```python
x = [x * 0.1 for x in range(-100, 100)]
senx = [math.sin(xs) for xs in x]
cosx = [math.cos(xs) for xs in x]

fig = plt.figure()
plt.plot(x, senx, 'r-', label="seno")
plt.plot(x, cosx, 'b--', label="cosseno");
plt.legend(loc=2)
plt.show()
```

```python
plt.plot(x, [y-1 for y in senx], color='blue')        # specify color by name
plt.plot(x, [y-2 for y in senx], color='g')           # short color code (rgbcmyk)
plt.plot(x, [y-3 for y in senx], color='0.75')        # Grayscale between 0 and 1
plt.plot(x, [y-4 for y in senx], color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, [y-5 for y in senx], color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, [y-6 for y in senx], color='chartreuse'); # all HTML color names supported
plt.show()
```

```python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, senx)

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, cosx);
```

```python
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
notas = [83,95,91,87,70,3,85,82,92,67,73,77,11]

plt.hist(notas, bins=10, range=(0,100))

plt.xlabel("Notas")
plt.show()
```

```python
plt.hist(notas, bins=10, range=(0,100))

# x-axis labels at 0, 10, ..., 100
plt.xticks([10 * i for i in range(11)])

plt.xlabel("Notas")
plt.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(1, figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.show()
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```

```python
np.random.seed(19680801)
data = np.random.randn(2, 100)

fig, axs = plt.subplots(2, 2, figsize=(5, 5))
axs[0, 0].hist(data[0])
axs[1, 0].scatter(data[0], data[1])
axs[0, 1].plot(data[0], data[1])
axs[1, 1].hist2d(data[0], data[1])

plt.show()
```

### Mais exemplos

* Exemplos de plots do matplotlib: https://matplotlib.org/tutorials/introductory/sample_plots.html

* Podem ser usadas diversas outras bibliotecas para plotar gráficos, entre elas, a `seaborn` (http://seaborn.pydata.org/), construída a partir da `matplotlib`, mas permitindo a produção de visualizações mais bonitas (e mais complexas)

```python
sb.set(style="white")

# Load the example mpg dataset
mpg = sb.load_dataset("mpg")

# Plot miles per gallon against horsepower with other semantics
sb.relplot(x="horsepower", y="mpg", hue="origin", size="weight",
           sizes=(40, 400), alpha=.5, palette="muted",
           height=6, data=mpg)
```

```python
sb.set()

# Load the example iris dataset
planets = sb.load_dataset("planets")

cmap = sb.cubehelix_palette(rot=-.2, as_cmap=True)
ax = sb.scatterplot(x="distance", y="orbital_period",
                    hue="year", size="mass",
                    palette=cmap, sizes=(10, 200),
                    data=planets)
```

```python
sb.set(style="white")

df = sb.load_dataset("iris")

g = sb.PairGrid(df, diag_sharey=False)
g.map_lower(sb.kdeplot)
g.map_upper(sb.scatterplot)
g.map_diag(sb.kdeplot, lw=3)
```
